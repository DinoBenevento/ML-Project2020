## Import

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
sns.set_style("darkgrid")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import log_loss
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
```

## Lettura da csv

```python
#range(1,9) a causa di uno spazio all'inizio di ogni riga
monk1_train = pd.read_csv("data/monks-1.train", delimiter=" ", header=None, usecols=range(1,9),
                         names=['class', 'a1', 'a2', 'a3', 'a4', 'a5', 'a6', 'Id']) 

X = monk1_train.drop(columns = ['class', 'Id'])
y = monk1_train['class']
```

## Preprocessing

### P1. Rimpiazzamento dei valori NaN con la media
##### DOMANDE: 
1. Conviene utilizzare un'altra metrica anziché la media?
2. Come comportarsi con i campi che hanno subito l'encoding? (penso convenga allo stesso modo).

```python
X = X.fillna(X.mean())
```

### P2. Normalizzazione

```python
X = MinMaxScaler().fit_transform(X)

pd.DataFrame(X)
```

## KNN e RandomizedSearchCV

```python
#!# leafSizes = np.empty(110) #(per ball_tree e kd_tree), default 30
#!# np.put(leafSizes,  np.arange(110),  np.arange(1, 111)) # da 10 a 110

#p = range(1, 20)
#nn = range(1, 31)
# vedere anche i parametri: metric_params, n_jobs
# metric = 'minkowski' di default 
```

```python
rs = RandomizedSearchCV(KNeighborsClassifier(metric = 'minkowski', leaf_size = 30), { 
    'n_neighbors': range(1, 50),
    'weights': ['uniform', 'distance'],
    'algorithm': ['ball_tree', 'kd_tree', 'brute'],
    'p': range(1, 10)
}, cv=5, return_train_score=True, n_iter=150)

rs.fit(X, y)
results = rs.cv_results_

resDf = pd.DataFrame(results)
resDf
```

```python
redResDf = resDf[['param_n_neighbors', 'param_weights','param_algorithm', 'param_p', 'mean_train_score', 'mean_test_score']]
redResDf
```
